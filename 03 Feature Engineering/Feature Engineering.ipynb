#### Step 4 - Data cleaning / Handling the missing Values
1. Categorical (text)- mode(most frequently)
2. COntinuous (Numerical )- Mean/Meadian

```python
X.columns
```

```python
X.dtypes
```

```python
categorical_cols = list(X.columns[X.dtypes == "object"])
categorical_cols
```

```python
continuous_cols = list(X.columns[X.dtypes != "object"])
continuous_cols
```

```python
df["AirBags"].mode()
```

```python
df["AirBags"].value_counts()
```

```python
 df["Luggage.room"].median()
```

```python
df["Luggage.room"].mean()
```

```python
def replacer(df):
    cat = list(df.columns[df.dtypes == "object"])
    for col in df.columns:
        if col in cat:
            mode = df[col].mode()[0]
            df[col] = df[col].fillna(mode)
        else:
            median = df[col].median()
            df[col] = df[col].fillna(median)

    print("missing values replaced")
```

```python
X.isna().sum()
```

```python
replacer(X)
```

```python
X.isna().sum()
```

```python
X.to_csv("Cleaned.csv", index=False)
```

#### Step - 5 Prepossing
1. Categorical - OneHOtEncoding
2. Continuous - StandardScaler

```python
categorical_cols
```

```python
continuous_cols
```

```python
X_cat = X[categorical_cols]
X_cat
```

```python
X_con = X[continuous_cols]
X_con
```

### Apply OneHotENcoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(
    handle_unknown="ignore", sparse_output=False, drop="first"
).set_output(transform="pandas")
X_cat_pre = ohe.fit_transform(X_cat)
```

```python
X_cat_pre.head()
```

#### Continuous cols use StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().set_output(transform="pandas")

X_con_pre = scaler.fit_transform(X_con)
X_con_pre
```

### Combining X_cat_pre and X_con_pre

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```

### Step 5 - Apply train test split

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2)
```

### Step 7 - Model Evaluation

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(xtrain, ytrain)
```

### Step 8 - Model evaluation

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score,
)
```

```python
def evaluate_model(model, x, y):
    # Predict the result for x
    ypred = model.predict(x)

    # Evaluate different metrics
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)

    # Print the metrics
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
print("Training Results : ")
evaluate_model(model, xtrain, ytrain)
```

```python
print("Testing Results : ")
evaluate_model(model, xtest, ytest)
```
